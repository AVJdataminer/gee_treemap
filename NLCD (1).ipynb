```python
import os
import zipfile
from google.colab import drive

# Step 1: Mount Google Drive
#drive.mount('/content/drive')

# Step 2: Define the source directory and list of states
source_dir = '/content/drive/MyDrive/FireWatch/nlcd'
states = [
    'Arkansas', 'Colorado', 'Connecticut', 'Delaware', 'Illinois', 'Indiana', 'Iowa',
    'Kansas', 'Kentucky', 'Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan',
    'Minnesota', 'Mississippi', 'Missouri', 'Montana', 'Nebraska', 'New Hampshire',
    'New Jersey', 'New Mexico', 'New York', 'North Carolina', 'North Dakota', 'Ohio',
    'Oklahoma', 'Pennsylvania', 'Rhode Island', 'South Carolina', 'South Dakota',
    'Tennessee', 'Vermont', 'Virginia', 'West Virginia', 'Wisconsin', 'Wyoming'
]

# Step 3: Define the function to zip each state folder
def zip_state_folders(source_dir, states):
    """
    Zips each specified state folder in the source directory for downloading to local machine.

    Parameters:
    source_dir (str): The path to the directory containing the state folders.
    states (list): List of state names corresponding to subdirectory names.
    """
    for state in states:
        # Construct the full path to the state folder
        state_folder_path = os.path.join(source_dir, state)
        if not os.path.exists(state_folder_path):
            print(f"State folder not found: {state_folder_path}")
            continue

        # Path for the zip file
        zip_path = f"/content/{state}.zip"

        # Zip the folder
        with zipfile.ZipFile(zip_path, 'w') as zipf:
            for root, _, files in os.walk(state_folder_path):
                for file in files:
                    file_path = os.path.join(root, file)
                    zipf.write(file_path, os.path.relpath(file_path, state_folder_path))

        print(f"{state} folder zipped successfully as {zip_path}")

# Step 4: Call the function to zip each state folder
zip_state_folders(source_dir, states)

# Step 5: Download the zipped files
print("All state folders are zipped and ready for download from the Colab file explorer.")
```

```python
def delete_tiles(state_name):
    """
    Deletes all files in the current directory and its subdirectories that start with
    'Wyoming_nlcd_2021_tile' but keeps 'Wyoming_nlcd_merged' files.

    Parameters:
    state_name (str): The state name prefix for the files (default is 'Wyoming').

    Returns:
    None
    """
    # Define the pattern to match for the tile files
    tile_prefix = f"{state_name}_nlcd_2021_tile"
    keep_prefix = f"{state_name}_nlcd_merged"

    # Traverse the current working directory and its subdirectories
    for root, dirs, files in os.walk(os.getcwd()):
        for file in files:
            # Check if the file starts with the tile prefix and doesn't match the keep prefix
            if file.startswith(tile_prefix) and not file.startswith(keep_prefix):
                # Construct full file path
                file_path = os.path.join(root, file)
                try:
                    # Remove the file
                    os.remove(file_path)
                    print(f"Deleted file: {file_path}")
                except Exception as e:
                    print(f"Failed to delete {file_path}: {e}")

# usage:
delete_tiles(state_name= 'Texas')
```

```python
def clean_processed_state_folders():
    """
    Iterates through each processed state folder, deletes the original merged .tif file
    if the reprojected (_4326.tif) file exists, and keeps only the projected files.
    """
    # Get the current working directory
    root_folder = os.getcwd()

    # Iterate over each directory in the current working directory
    for state_folder in os.listdir(root_folder):
        state_folder_path = os.path.join(root_folder, state_folder)

        # Ensure it's a directory
        if os.path.isdir(state_folder_path):
            # Look for the merged .tif and the reprojected _4326.tif in the folder
            for file in os.listdir(state_folder_path):
                if file.endswith("_merged.tif"):
                    original_tif_path = os.path.join(state_folder_path, file)
                    reprojected_tif_path = os.path.join(state_folder_path, file.replace("_merged.tif", "_merged_4326.tif"))

                    # Delete the original merged .tif if the reprojected file exists
                    if os.path.exists(reprojected_tif_path):
                        try:
                            os.remove(original_tif_path)
                            print(f"Deleted original merged file: {original_tif_path}")
                        except Exception as e:
                            print(f"Failed to delete {original_tif_path}: {e}")
                    else:
                        print(f"No reprojected file found for {original_tif_path}, keeping the original.")

clean_processed_state_folders()
```

```python
#this one should check for tiles that have already been downloaded in a previous run
import os
import ee
import geemap
import rasterio
from rasterio.merge import merge
from rasterio.warp import calculate_default_transform, reproject, Resampling

ee.Authenticate()
ee.Initialize(project='nlcd-project')

# Initialize the Earth Engine API
ee.Initialize()

# Function to create a folder for the state's tiles
def create_state_folder(state_name):
    folder_path = os.path.join(os.getcwd(), state_name)
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
    return folder_path

# Function to count the number of downloaded tiles
def get_downloaded_tiles_count(state_name):
    folder_path = os.path.join(os.getcwd(), state_name)
    tile_files = [f for f in os.listdir(folder_path) if f.endswith('.tif')]
    return len(tile_files)

# Function to export landcover tiles starting from where it left off
def export_landcover_tiles(state_name):
    # Define the region of interest using the state name
    state = ee.FeatureCollection("TIGER/2018/States") \
              .filter(ee.Filter.eq('NAME', state_name))

    # Load the NLCD dataset and filter for the 2021 product
    dataset = ee.ImageCollection('USGS/NLCD_RELEASES/2021_REL/NLCD') \
                .filter(ee.Filter.eq('system:index', '2021')) \
                .first()

    # Select the landcover band
    landcover = dataset.select('landcover')

    # Split the region into smaller tiles (20x20 grid for Texas)
    grid = geemap.fishnet(state, rows=20, cols=20)

    # Create a folder for the state to save the tiles
    folder_path = create_state_folder(state_name)

    # Get the count of already downloaded tiles to resume from where it left off
    start_index = get_downloaded_tiles_count(state_name)

    # Loop over each grid tile, skipping those already downloaded
    for i, feature in enumerate(grid.getInfo()['features']):
        if i < start_index:
            print(f"Skipping tile {i+1} for {state_name}, already downloaded.")
            continue

        tile = ee.Feature(feature)
        tile_geometry = tile.geometry()
        output_tif = os.path.join(folder_path, f'{state_name}_nlcd_2021_tile_{i+1}.tif')

        # Export each tile separately
        try:
            geemap.ee_export_image(
                landcover.clip(tile_geometry),
                filename=output_tif,
                scale=30,  # Set to 30m to match NLCD resolution
                region=tile_geometry,
                file_per_band=False
            )
            print(f"Exported tile {i+1} for {state_name} to {output_tif}")
        except Exception as e:
            print(f"Failed to export tile {i+1} for {state_name}: {e}")

# Remaining functions: load_tiles_from_state_folder, merge_tiles, reproject_geotiff, delete_tiles,
# merge_state_tiles, and process_multiple_states remain unchanged.
# Function to load all the GeoTIFF tiles for a specific state
def load_tiles_from_state_folder(state_name):
    # Path to the state folder
    folder_path = os.path.join(os.getcwd(), state_name)

    # Find all .tif files in the state folder
    tile_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".tif") and 'merged' not in file:
                tile_files.append(os.path.join(root, file))

    # Open each tile and add to the list
    src_files_to_mosaic = []
    for tile in tile_files:
        src = rasterio.open(tile)
        src_files_to_mosaic.append(src)

    return src_files_to_mosaic, folder_path, tile_files

# Function to merge the tiles into one GeoTIFF
def merge_tiles(src_files_to_mosaic, output_path):
    # Perform the merging
    mosaic, out_trans = merge(src_files_to_mosaic)

    # Get metadata of one of the input files for reference
    out_meta = src_files_to_mosaic[0].meta.copy()

    # Update the metadata with the merged file's dimensions, transform, and CRS
    out_meta.update({
        "driver": "GTiff",
        "height": mosaic.shape[1],
        "width": mosaic.shape[2],
        "transform": out_trans,
        "crs": src_files_to_mosaic[0].crs
    })

    # Write the merged file to the output path
    with rasterio.open(output_path, "w", **out_meta) as dest:
        dest.write(mosaic)

    print(f"Merged GeoTIFF saved to {output_path}")

# Function to reproject the merged GeoTIFF to EPSG:4326 and save it with _4326 appended to the filename
def reproject_geotiff(input_tif_path, target_crs='EPSG:4326'):
    """
    Reprojects the given GeoTIFF file to a target CRS and saves it with '_4326' appended to the filename.

    Parameters:
    input_tif_path (str): Path to the input GeoTIFF file.
    target_crs (str): The target CRS in EPSG format (default is 'EPSG:4326').

    Returns:
    str: The path to the reprojected GeoTIFF file.
    """
    # Derive the output file path by appending '_4326' to the original filename
    base, ext = os.path.splitext(input_tif_path)
    output_tif_path = f"{base}_4326{ext}"

    with rasterio.open(input_tif_path) as src:
        # Calculate the transform and dimensions for the new projection
        transform, width, height = calculate_default_transform(
            src.crs, target_crs, src.width, src.height, *src.bounds)

        # Update metadata with the new projection details
        kwargs = src.meta.copy()
        kwargs.update({
            'crs': target_crs,
            'transform': transform,
            'width': width,
            'height': height
        })

        # Reproject the data and save to a new GeoTIFF
        with rasterio.open(output_tif_path, 'w', **kwargs) as dst:
            for i in range(1, src.count + 1):
                reproject(
                    source=rasterio.band(src, i),
                    destination=rasterio.band(dst, i),
                    src_transform=src.transform,
                    src_crs=src.crs,
                    dst_transform=transform,
                    dst_crs=target_crs,
                    resampling=Resampling.nearest
                )

        print(f"Reprojection complete. GeoTIFF saved as {output_tif_path}")

    return output_tif_path

# Function to delete individual tiles
def delete_tiles(tile_files):
    for tile_file in tile_files:
        try:
            os.remove(tile_file)
            print(f"Deleted tile: {tile_file}")
        except Exception as e:
            print(f"Failed to delete {tile_file}: {e}")

# Main function to handle the merging process, reprojection, and deletion of tiles
def merge_state_tiles(state_name):
    # Load all tiles from the state's directory
    src_files_to_mosaic, folder_path, tile_files = load_tiles_from_state_folder(state_name)

    # Define the output merged GeoTIFF path
    output_tiff = os.path.join(folder_path, f'{state_name}_nlcd_merged.tif')

    # Merge the tiles and save as a single GeoTIFF
    merge_tiles(src_files_to_mosaic, output_tiff)

    # Reproject the merged GeoTIFF to EPSG:4326 and save it with _4326 appended
    reprojected_tiff = reproject_geotiff(output_tiff)

    # Delete individual tile files after merging
    delete_tiles(tile_files)

# Function to process multiple states
def process_multiple_states(state_names):
    for state_name in state_names:
        print(f"Processing state: {state_name}")

        # Export landcover tiles for the state
        export_landcover_tiles(state_name)

        # Merge the exported tiles into a single GeoTIFF, reproject, and delete the individual tiles
        merge_state_tiles(state_name)
        print(f"Finished processing state: {state_name}\n")


# Example list of state names to process
state_names = ['Texas']  # Replace with the list of states you want to process

# Process all the states in the list
process_multiple_states(state_names)
```

# Reproject the merged tifs

```python
# Function to reproject the merged GeoTIFF to EPSG:4326 and save it with _4326 appended to the filename
def reproject_geotiff(input_tif_path, target_crs='EPSG:4326'):
    """
    Reprojects the given GeoTIFF file to a target CRS and saves it with '_4326' appended to the filename.

    Parameters:
    input_tif_path (str): Path to the input GeoTIFF file.
    target_crs (str): The target CRS in EPSG format (default is 'EPSG:4326').

    Returns:
    str: The path to the reprojected GeoTIFF file.
    """
    # Derive the output file path by appending '_4326' to the original filename
    base, ext = os.path.splitext(input_tif_path)
    output_tif_path = f"{base}_4326{ext}"

    with rasterio.open(input_tif_path) as src:
        # Calculate the transform and dimensions for the new projection
        transform, width, height = calculate_default_transform(
            src.crs, target_crs, src.width, src.height, *src.bounds)

        # Update metadata with the new projection details
        kwargs = src.meta.copy()
        kwargs.update({
            'crs': target_crs,
            'transform': transform,
            'width': width,
            'height': height
        })

        # Reproject the data and save to a new GeoTIFF
        with rasterio.open(output_tif_path, 'w', **kwargs) as dst:
            for i in range(1, src.count + 1):
                reproject(
                    source=rasterio.band(src, i),
                    destination=rasterio.band(dst, i),
                    src_transform=src.transform,
                    src_crs=src.crs,
                    dst_transform=transform,
                    dst_crs=target_crs,
                    resampling=Resampling.nearest
                )

        print(f"Reprojection complete. GeoTIFF saved as {output_tif_path}.")

    return output_tif_path
```
